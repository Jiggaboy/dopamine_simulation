# Directional Spatial Clustering
The idea is fairly simple and similar to other appraches, e.g. OPTICS? or ST-DBSCAN. The key difference here is that clusters are only formed in feedforward manner.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
import pathlib
from sklearn.cluster import DBSCAN
```

```python
file = pathlib.Path("./sample_spike_train.npy")
arr = np.load(file)
```

```python
fig = plt.figure()  
ax = fig.add_subplot(111, projection='3d')  
ax.scatter(*arr.T)
plt.show()
```
