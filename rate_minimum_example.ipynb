# Minimum example for a sigmoid-rate model
## Question: Why is the rate of the neuron not bounded between 0 and gain?

```python
import nest
nest.ResetKernel()
neuron = nest.Create('sigmoid_rate_ipn')
neuron.set(g=1.)
```

```python
mm = nest.Create('multimeter', params={'record_from': ['rate'],
                                       'interval': 1.})
nest.Connect(mm, neuron)
nest.Simulate(1000.0)
print(neuron.get("rate"))
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(mm.events["rate"])
```
